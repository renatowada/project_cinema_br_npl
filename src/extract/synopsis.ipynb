```python
import json
import os
import unicodedata

# Função auxiliar para remover acentos e minúsculas
def normalize_title(text):
    if not isinstance(text, str): return ""
    return unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8').lower().strip()


path_master = '../data/movies_master.json'
path_output = '../data/synopsis.json'


if not os.path.exists(path_master):
    raise FileNotFoundError(f"Arquivo mestre não encontrado em {path_master}. Rode o data.ipynb primeiro!")

with open(path_master, 'r', encoding='utf-8') as f:
    master_list = json.load(f)


# Chave = Título Normalizado, Valor = Sinopse
user_synopsis_map = {}

if 'highlights_synopsis' not in locals():
    raise NameError("A lista 'highlights_synopsis' não foi encontrada. Corrija os erros de sintaxe e rode a célula da lista primeiro.")

for item in highlights_synopsis:
    # Usa o título normalizado como chave
    clean_title = normalize_title(item.get('title'))
    if clean_title:
        user_synopsis_map[clean_title] = item.get('synopsis', "")

final_synopsis_list = []
missing_count = 0

for movie in master_list:
    original_title = movie['title']
    clean_title_master = normalize_title(original_title)
    
    synopsis_text = user_synopsis_map.get(clean_title_master, "")
    
    if not synopsis_text:
        synopsis_text = user_synopsis_map.get(normalize_title(movie.get('original_title_raw', '')), "")
    
    if not synopsis_text:
        missing_count += 1


    entry = {
        "title": original_title,          # Nome oficial da master
        "release_year": movie['release_year'], # Ano oficial da master
        "synopsis": synopsis_text
    }
    final_synopsis_list.append(entry)

os.makedirs(os.path.dirname(path_output), exist_ok=True)
with open(path_output, 'w', encoding='utf-8') as f:
    json.dump(final_synopsis_list, f, ensure_ascii=False, indent=4)

print(f"PROCESSAMENTO CONCLUÍDO!")
print(f"- Filmes processados: {len(final_synopsis_list)}")
print(f"- Sinopses encontradas: {len(final_synopsis_list) - missing_count}")
print(f"- Sinopses vazias: {missing_count}")
print(f"- Arquivo salvo em: {os.path.abspath(path_output)}")
```

```python
import pandas as pd
import json
import unidecode
import re
import os


def generate_id_movie(title, release_year):

    title = str(title) if pd.notnull(title) else ""
    

    try:
        release_year = str(int(release_year))
    except (ValueError, TypeError):
        release_year = "0000"
    
    clear_title = title.lower().strip()
    clear_title = unidecode.unidecode(clear_title)
    clear_title = re.sub(r'[^a-z0-9]', '', clear_title)
    
    return f"{clear_title}_{release_year}"


json_path = '../data/synopsis.json'

with open(json_path, 'r', encoding='utf-8') as f:
    data_list = json.load(f)


df = pd.DataFrame(data_list)


df['id_filme'] = df.apply(lambda row: generate_id_movie(row['title'], row['release_year']), axis=1)


print(df.head())
```
